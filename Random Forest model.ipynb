```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('exp_train.csv')
df
```

```python
df = df.drop(['Customer_ID', 'Month', 'Occupation', 'Payment_of_Min_Amount', 'Payment_Behaviour'], axis=1)
df
```

```python
df.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier(max_depth=None, random_state=2)
```

```python
X = df.drop(['Credit_Score'], axis = 1)
y = df['Credit_Score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
rfc_model.fit(X_train, y_train)
```

```python
y_test_pred = rfc_model.predict(X_test)
y_test_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, log_loss, classification_report

accuracy = accuracy_score(y_test, y_test_pred)
accuracy
```

```python
y_test_pred_proba = rfc_model.predict_proba(X_test)

logloss = log_loss(y_test, y_test_pred_proba)
logloss
```

```python
cm = confusion_matrix(y_test, y_test_pred)
cm
```

```python
print(classification_report(y_test, y_test_pred))
```
