# Data Cleaning

Identify problems
- ID, Name, SSN columns need to be removed (irrelevant for analysis)
- Age, Occupation, Num_of_Loan, Num_of_Delayed_Payment, Changed_Credit_Limit, Credit_Mix, Amount_invested_monthly have "_"
- Age, Annual_Income, Num_of_Loan, Num_of_Delayed_Payment, Changed_Credit_Limit, Credit_Mix, Outstanding_Debt, Amount_invested_monthly, Monthly_Balance have wrong data type
- Cleaning each columns

```python
#Delete ID, Name, SSN columns
df = df.drop(['ID', 'Name', 'SSN'], axis = 1)
df
```

```python
#delete "_"

columns_task2 = ['Age', 'Annual_Income', 'Changed_Credit_Limit', 'Num_of_Delayed_Payment', 'Credit_Mix', 'Occupation']

for col in columns_task2: 
    df[col] = df[col].str.replace('_', '')
    
df['Annual_Income'][10]
```

```python
#change data type

def change_type(i):
    try:
        return float(i)
    except ValueError:
        return i

col_change_type = ['Age', 'Annual_Income', 'Num_of_Loan', 'Num_of_Delayed_Payment', 
                   'Changed_Credit_Limit', 'Amount_invested_monthly', 'Outstanding_Debt', 'Monthly_Balance']

for col in col_change_type:
    df[col] = pd.to_numeric(df[col], errors='coerce').apply(change_type)
```

```python
df.dtypes
```

```python
#Age(value > 110, negative values)

df['Age'] = df['Age'].mask((df['Age'] > 110) | (df['Age'] < 0), np.nan)
df['Age']
```

```python
#Type_of_Loan(create new 8 columns for 8 loan type, drop column Type_of_Loan)

for i in df['Type_of_Loan'].value_counts().head(9).index[1:] : 
    df[i] = df['Type_of_Loan'].str.contains(i)
    
df = df.drop('Type_of_Loan', axis = 1)
df
```

```python
#Change datatype for new 8 columns

for col in df.columns[-8:]:
    df[col] = df[col].astype(float)
```

```python
#Num_Bank_Accounts(drop negative values)

df['Num_Bank_Accounts'] = df['Num_Bank_Accounts'].mask(df['Num_Bank_Accounts'] < 0, np.nan)
(df['Num_Bank_Accounts']< 0).sum()
```

```python
#Credit_History_Age

import re

def History_age(age):
    try : 
        years = int("".join(re.findall('[0-9]',''.join(age.split("and")[0]))))
        month = int("".join(re.findall('[0-9]',''.join(age.split("and")[1]))))
        return years*12 + month
    except :
        return np.nan
    
df['Credit_History_Age'] = df['Credit_History_Age'].apply(History_age)
```

```python
df['Credit_History_Age']
```

```python
#Payment_of_Min_Amount

df['Payment_of_Min_Amount'] = df['Payment_of_Min_Amount'].str.replace('NM', 'No')
df['Payment_of_Min_Amount'].value_counts()
```

```python
#Payment_Behaviour(delete the '!@9#%8' values)

df.replace("!@9#%8", np.nan, inplace=True)
df['Payment_Behaviour'].value_counts()
```

# Data Preprocessing

```python
#Credit_Mix

cm = {'Good': 2, 'Standard': 1, 'Bad': 0, '': np.nan}
df['Credit_Mix'] = df['Credit_Mix'].map(cm)
df['Credit_Mix']
```

```python
#Credit_Score

df['Credit_Score'].value_counts()

cs = {'Good': 2, 'Standard': 1, 'Poor': 0}
df['Credit_Score'].map(cs)
```

```python
#See how many missing values

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    miss = '{} - {}%'.format(col, pct_missing)
    print(miss)
```

```python
for col in df.select_dtypes(exclude='object'):
    df[col] = df[col].fillna(df[col].mean())
```

```python
#fill the Payment_Behaviour object columns using SimpleImputer
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="most_frequent")
imputer.fit(df[['Payment_Behaviour']])
df[['Payment_Behaviour']] = imputer.transform(df[['Payment_Behaviour']])
```
